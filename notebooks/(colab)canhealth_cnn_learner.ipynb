```python
from fastai import *
from fastai.vision import *
from fastai.vision.all import *
from fastai.vision.all import imagenet_stats, ClassificationInterpretation
from fastai.metrics import error_rate, accuracy
from fastai.vision.data import ImageDataLoaders
import gdown
```

```python
from fastai import get_transforms
```

```python
# prompt: import imagenet_stats, ClassificationInterpretation, get_transforms

from fastai.vision.learner import *
from fastai.vision.models.xresnet import *
from fastai.vision.models.unet import *
from fastai.callback.schedule import *
from fastai.callback.hook import *
from fastai.callback.progress import *
from fastai.callback.tracker import *


from fastai.vision.all import imagenet_stats, ClassificationInterpretation, get_transforms
```

```python
# Defining the path to the data
data_path = '/content/cannabis-health-diagnosis/data'
```

```python
# Reading the image-based dataset
img_data = ImageDataLoaders.from_folder(path=data_path, train='train', valid='valid', ds_tfms=get_transforms(), size=224, bs=100)
```

```python
img_data.normalize(imagenet_stats)
```

```python
img_data.show_batch(rows=4, figsize=(12,10))
```

```python
model = cnn_learner(img_data, models.resnet34, metrics=[accuracy, error_rate])
```

```python
model.summary()
```

```python
model.fit_one_cycle(7)
```

```python
model.fit_one_cycle(5)
```

```python
model.save('models/cnn_learner_12cycles')
```

```python
model.save('modelsh5/cnn_learner_12cycles.h5')
```

```python
interpret = ClassificationInterpretation.from_learner(model)
```

```python
interpret.plot_top_losses(4, figsize=(20, 25))
```

```python
interpret.plot_confusion_matrix(figsize=(20,20), dpi=60)
```

```python
model.lr_find()
```

```python
model.recorder.plot()
```

```python
model.unfreeze()
model.fit_one_cycle(3, max_lr=slice(1e-03, 1e-02))
```

```python
model.unfreeze()
model.fit_one_cycle(5, max_lr=slice(1e-03, 1e-02))
```

```python
model.save('models/cnn_learner_lr8cycles')
```

```python
model.save('modelsh5/cnn_learner_lr8cycles.h5')
```

```python
model.freeze()
model.lr_find()
model.recorder.plot()
```

```python
lr = 1e-3/2
model.fit_one_cycle(2, slice(lr))
```

```python
model.fit_one_cycle(3, slice(lr))
```

```python
model.save('models/cnn_learner_final5')
```

```python
model.save('modelsh5/cnn_learner_final5.h5')
```
